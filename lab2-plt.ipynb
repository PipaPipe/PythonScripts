```python
import matplotlib.pyplot as plt
```

```python
import os

PATH = "/Users/maksimsadkov/Downloads/test/"


def change_name(name):
    return name.split()[1][:-4].replace("D", "e")


def exp_to_decimal(exp):
    exp = exp.replace("D", "e")
    dec_part = exp[:exp.index("e")]
    exp_part = exp[exp.index("e")+1:]
    return float(dec_part) * 10**int(exp_part)


def get_max_value(max_value, comparable_value):
    if max_value == 0:
        max_value = comparable_value
    if max_value < comparable_value:
        max_value = comparable_value
    return max_value


def get_file_content(path):
    with open(path) as f:
        return f.readlines()


list_dir = os.listdir(PATH)
max_value = 0
for i in list_dir:
        content = get_file_content(os.path.join(PATH, i))
        max_value = 0
        for j in content:
            n = j.split()
            print(n[1], exp_to_decimal(n[1]))
            max_value = get_max_value(max_value, exp_to_decimal(n[1]))
        print(i)
        print(change_name(i), max_value, sep=" — ")
```

```python
# Выкинем крайние значения
def round_list(list, value):
    return list[value:][:-value]
```

```python
# Функция создания графика
def create_plot(title, x, y, y1):
    fig, ax = plt.subplots()  # Create a figure containing a single axes.    
    plt.title(title)
    plt.xlabel("t")
    plt.ylabel("Ψ")
    plt.plot(x, y)  # Plot some data on the axes.
    plt.plot(x, y1)
```

```python
# получим числа из названий файлов
def get_names(list_dir):
    result = set()
    for i in list_dir:
        result.add(round(exp_to_decimal(change_name(i))))
    return list(result)
    
list_res = get_names(list_dir)
list_res
list_res
```

```python
# получим словарь соответствий файлов
def get_dict_psi_to_psi2(list_dir):
    psi_to_psi2_dict = {}
    for i in list_dir:
        if "psi_G" in i:
            psi_to_psi2_dict[i] = i.replace("psi", "psi2")
    return psi_to_psi2_dict

psi_to_psi2_dict = get_dict_psi_to_psi2(list_dir)
psi_to_psi2_dict
```

```python
# функция взятия максимального числа из половины списка
def get_max_of_half(lst):
    return max(lst[int(len(lst)/2):])

psi2_max_of_half = []
```

```python
def clear_lists(*lists):
    return list(map(list.clear, lists))
```

```python
first_column = []
second_column = []
psi2_max = []
for i, val in enumerate(list_dir):
    first_column = []
    second_column = []
    psi2_max = []
    if "psi_G" in val:
        content = get_file_content(os.path.join(PATH, val))
        for j in content:
            n = j.split()
            first_column.append(exp_to_decimal(n[0]))
            second_column.append(exp_to_decimal(n[1]))
            
        content = get_file_content(os.path.join(PATH, psi_to_psi2_dict[val]))
        for k in content:
            n = k.split()
            psi2_max.append(exp_to_decimal(n[1]))
        
        psi2_max_of_half.append(get_max_of_half(psi2_max))
        
        first_column = round_list(first_column, 100)
        second_column = round_list(second_column, 100)
        psi2_max = round_list(psi2_max, 100)
        
        
        create_plot(val, first_column, second_column, psi2_max)
        
    
```

```python
psi2_max_of_half.sort()
plt.plot(list_res, psi2_max_of_half)
```

```python
import unittest
```

```python
class TestLab1Functions(unittest.TestCase):
    def test_change_name(self):
        result = change_name('psi2_G= 0.172500D+04.dat')
        self.assertEqual(result, '0.172500e+04')
        self.assertIsInstance(result, str)
        
    def test_exp_to_decimal(self):
        result = exp_to_decimal('0.475919D-01')
        self.assertEqual(result, 0.0475919)
        self.assertIsInstance(result, float)
        
    def test_get_max_value(self):
        self.assertEqual(get_max_value(4, 5), 5)
        self.assertEqual(get_max_value(0, 0), 0)
        self.assertIsInstance(get_max_value(4, 5), int)
        
    def test_get_names(self):
        PATH = "/Users/maksimsadkov/Downloads/test/"
        list_dir = os.listdir(PATH)
        result = get_names(list_dir)
        self.assertEqual(result, [1730, 1740, 1710, 1715, 1750, 1720, 1725])
    
    def test_get_dict_psi_to_psi2(self):
        PATH = "/Users/maksimsadkov/Downloads/test/"
        list_dir = os.listdir(PATH)
        result = get_dict_psi_to_psi2(list_dir)
        self.assertIsInstance(result, dict)
        self.assertEqual(result, {'psi_G= 0.173000D+04.dat': 'psi2_G= 0.173000D+04.dat',
                                     'psi_G= 0.172500D+04.dat': 'psi2_G= 0.172500D+04.dat',
                                     'psi_G= 0.171000D+04.dat': 'psi2_G= 0.171000D+04.dat',
                                     'psi_G= 0.174000D+04.dat': 'psi2_G= 0.174000D+04.dat',
                                     'psi_G= 0.172000D+04.dat': 'psi2_G= 0.172000D+04.dat',
                                     'psi_G= 0.171500D+04.dat': 'psi2_G= 0.171500D+04.dat',
                                     'psi_G= 0.175000D+04.dat': 'psi2_G= 0.175000D+04.dat'}
                        )
```
