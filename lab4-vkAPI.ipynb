```python
def get_friends(access_token, fields, user_id):
    url = 'https://api.vk.com/method/'+ "friends.get"+\
    '?access_token='+ access_token+ '&user_id=' +\
    str(user_id) +'&fields='+ ','.join(fields)+'&v=5.154'
    response = requests.get(url).json()
    return response
```

```python
def post(access_token, message, owner_id):
    url = 'https://api.vk.com/method/'+ "wall.post"+\
    '?access_token='+ access_token + '&message=' + message +'&owner_id='+str(owner_id) +'&v=5.154'
    response = requests.get(url).json()
    return response
```
